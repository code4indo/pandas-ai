```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = dataframe
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Month_Year'] = df['Timestamp'].dt.to_period('M')
grouped_df = df.groupby(['Month_Year', 'Is Laundering']).size().reset_index(
    name='Count')
heatmap_df = grouped_df.pivot(index='Month_Year', columns='Is Laundering',
    values='Count')
sns.heatmap(heatmap_df, cmap='Blues', annot=True, fmt='g')
plt.title('Distribution of Transactions by Laundering Status')
plt.xlabel('Laundering Status')
plt.ylabel('Month-Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = dataframe
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Month_Year'] = df['Timestamp'].dt.strftime('%Y-%m')
monthly_counts = df.groupby('Month_Year')['Timestamp'].count()
colors = ['blue', 'green', 'orange', 'red', 'purple']
for i, year in enumerate(range(2022, 2027)):
    year_data = monthly_counts.loc[f'{year}-01':f'{year}-12']
    plt.bar(year_data.index, year_data.values, color=colors[i])
plt.xticks(rotation=45, ha='right')
plt.gca().xaxis.set_major_formatter(plt.FixedFormatter(monthly_counts.index))
plt.yticks(fontsize=8)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = dataframe
laundering_df = df[df['Is Laundering'] == 1]
num_laundering = len(laundering_df)
num_non_laundering = len(df) - num_laundering
labels = ['Laundering', 'Non-Laundering']
sizes = [num_laundering, num_non_laundering]
colors = ['#ff9999', '#66b3ff']
explode = 0.1, 0
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors=colors, explode=explode, labels=labels, autopct=
    '%1.1f%%', startangle=90)
centre_circle = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.axis('equal')
plt.tight_layout()
plt.show()
```
